```python
l = []
def remove_bracket(data):
    for d in data:
        k = d.split('[')[0].split()
        if len(k) == 1:
            l.append(k[0])
        elif len(k) == 2:
            l.append(k[0] + k[1])
        else:
            l.append(k[0] + k[1] + k[2])
```

```python
remove_bracket(hiv1['Number of deaths due to HIV/AIDS'])
```

```python
l
```

```python
hiv1['Number of deaths due to HIV/AIDS_1'] = l
```

```python
hiv1.head()
```

```python
hiv1['Number of deaths due to HIV/AIDS_1'].tail(40)
```

```python
h = hiv1['Number of deaths due to HIV/AIDS_1'][179].split(';')[1]
```

```python
hiv1['Number of deaths due to HIV/AIDS_1'].loc[hiv1['Number of deaths due to HIV/AIDS_1'].index == 179] = h
```

```python
hiv1['Number of deaths due to HIV/AIDS_1'].loc[hiv1['Number of deaths due to HIV/AIDS_1'].index == 179]
```

```python
hiv1.head()
```

```python
hiv1.drop('Number of deaths due to HIV/AIDS', axis=1, inplace=True)
```

```python
hiv1.head()
```

```python
hiv1['Number of deaths due to HIV/AIDS_1'] = hiv1['Number of deaths due to HIV/AIDS_1'].astype('int64')
```

```python
hiv1.info()
```

```python
hiv1.to_csv('HIV_1_cleaned.csv', index=False)
```
