```python
def remove_punctuation(text):
    import string
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

```python
df['Start-up'] = df['Start-up'].apply(remove_punctuation)
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
import numpy as np
```

```python
sw = stopwords.words('english')
np.array(sw)
```

```python
len(sw)
```

```python
def stop_words(text):
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    return " ".join(text)
```

```python
df['Start-up'] = df['Start-up'].apply(stop_words)
df.head(10)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()
```

```python
for data in df['Start-up']:
    for words in data.split():
        fdist[words.lower()] += 1
```

```python
fdist
```

```python
fdist.most_common(10)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemma(text):
    text = [lemmatizer.lemmatize(word) for word in text.split()]
    return " ".join(text)
```

```python
df['startup'] = df['Start-up'].apply(lemma)
```

```python
delhi = ['new delhi', 'south west delhi']
hyderabad = ['hyd'] 
bangalore = ['banglore']
patna = ['bihar']
hoshangabad = ['hoshangabad m.p.']
mumbai = ['navi mumbai']
ponda = ['ponda goa']
```

```python
def rename(city):
    if city in delhi:
        newcity = 'delhi'
    elif city in hyderabad:
        newcity = 'hyderabad'
    elif city in bangalore:
        newcity = 'bangalore'
    elif city in patna:
        newcity = 'patna'
    elif city in hoshangabad:
        newcity = 'hoshangabad'
    elif city in mumbai:
        newcity = 'mumbai'
    elif city in ponda:
        newcity = 'ponda'
    else:
        newcity = city
    return newcity
```

```python
df['City'] = df['City'].apply(rename)
```
