```python
pune_str = ['pune', 'mancher ,pune', 'pune, maharashtra', 'ruchitamagr412@gmail.com', 'i want to attend',
            'near dikshant international school']

mumbai_str = ['mumbai', 'kharghar, navi mumbai - 410210', 'm', 'navi mumbai', 'navi mumbai-410210', 'goregaon w, mumbai',
             'navi mumbai - 410210', 'mumbai suburban', 'navi mumbai, new panvel', 'mumbai city', 'vashi, navi mumbai',
             'mumbai/lucknow']

delhi_str = ['delhi', 'new delhi', 'sudershan park,new delhi', 'delhi ncr', 'delhi.', 'south west delhi', 'west delhi',
            'north west delhi', 'new delhi, india', 'newdelhi', 'delhi-ncr', 'shahdara, north east delhi', 'delhi/lucknow',
            'haryana ,delhi', 'delhi/ncr', 'ncr']

patna_str = ['patna', 'patna(bihar)', 'patna, bihar', 'patna , bihar', 'patna (bihar)', 'bihar']

noida_str = ['noida', 'greater noida', 'greater noida (west)', 'greater noida west']

bangalore_str = ['bengaluru', 'bangalore south', 'bangalore', 'banglore']

gurgaon_str = ['gurgaon', 'gurugram', 'gurgoan', 'gurgram']
```

```python
for (row, col) in df.iterrows():
    if str.lower(col.cities) in pune_str:
        df['cities'].replace(to_replace=col.cities, value= 'pune', inplace=True)
        
    if str.lower(col.cities) in mumbai_str:
        df['cities'].replace(to_replace=col.cities, value= 'mumbai', inplace=True)
        
    if str.lower(col.cities) in delhi_str:
        df['cities'].replace(to_replace=col.cities, value= 'delhi', inplace=True)
        
    if str.lower(col.cities) in patna_str:
        df['cities'].replace(to_replace=col.cities, value= 'patna', inplace=True)
        
    if str.lower(col.cities) in noida_str:
        df['cities'].replace(to_replace=col.cities, value= 'noida', inplace=True)
        
    if str.lower(col.cities) in bangalore_str:
        df['cities'].replace(to_replace=col.cities, value= 'bangalore', inplace=True)
        
    if str.lower(col.cities) in gurgaon_str:
        df['cities'].replace(to_replace=col.cities, value= 'gurgaon', inplace=True)    
```
